```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import math

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uniform_distribution = np.random.uniform(0, 1, 10000)
uniform_df = pd.DataFrame({"value": uniform_distribution})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=uniform_df)
```

```python
plt.figure(figsize=(10, 6))
sns.displot(uniform_df)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=stats.norm.pdf(uniform_df))
plt.show()
```

```python
mu = 0
variance = 1
sigma = np.sqrt(variance)
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 1000)
plt.figure(figsize=(10, 6))
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python
n = 100
p = 0.5

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
x = np.arange(stats.binom.ppf(0.01, n, p), stats.binom.ppf(0.99, n, p))
print(x)
print(stats.binom.ppf(0.01, n, p))
print(stats.binom.ppf(0.99, n, p))

ax.plot(x,  stats.binom.pmf(x, n, p), 'bo', ms=8, alpha=0.5)
ax.vlines(x, 0, stats.binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
rv = stats.binom(n, p)
plt.show()
```

```python
np.random.seed(1)

lognorm_dataset = stats.lognorm.rvs(s=.5, scale=math.exp(1), size=1000)

fig, axs =  plt.subplots(1, 2, figsize=(10, 4))

axs[0].hist(lognorm_dataset, bins=50, edgecolor='black')
stats.probplot(lognorm_dataset, plot=axs[1])

stats.kstest(lognorm_dataset, 'norm')
```

```python
#transform to normal distribution

np.random.seed(0)

data = np.random.beta(4, 15, 300)

data_log = np.cbrt(data)

fig, axs =  plt.subplots(2, 2, figsize=(10, 4))

axs[0][0].hist(data, bins=50, edgecolor='black')
axs[1][0].hist(data_log, bins=50, edgecolor='black')

stats.probplot(data, plot=axs[0][1])
stats.probplot(data_log, plot=axs[1][1])
```
